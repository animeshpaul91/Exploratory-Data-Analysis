### Chart -7 Flu Heat Map  of U.S.A

```r
statesData<-read.csv("Task4_Inputs/StateDataforMap_2018-19week4.csv")
statesData$region <- tolower(statesData$STATENAME)
library(ggplot2)
library(maps)
library(stringr)
library(usmap)
states <- map_data("state")
map.df <- merge(states,statesData, by="region", all.x=T)
map.df <- map.df[order(map.df$order),]
map.df$act<-as.integer(str_sub(map.df$ACTIVITY.LEVEL,-2,-1))

final<-data.frame(cbind(map.df$region,as.numeric(map.df$act)))
colnames(final)<-c("state","level")
us<-plot_usmap(data = final, values = "level", lines = "red") + 
  scale_fill_discrete(
    name = "Flu Level"
  ) + theme(legend.position = "right")
us+ggtitle("Heat Map Generated from CDC Flu Data of Season 2018-2019")+
theme(plot.title = element_text(hjust = 0.5,face="bold",size = 15))
```
