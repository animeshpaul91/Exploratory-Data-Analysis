## Problem 1

```r
sales1 <- c(12,14,16,29,30,45,19,20,16,19,34,20)
sales2 <- rpois(12,34) #Will return a 12 tupled vector with mean at 34 in Poisson Distribution
par(bg = "cornsilk")

plot(sales1, col = "blue", type = "o", ylim = c(0,100), xlab = "Month", ylab = "Sales") #By default, the count of 
#sales will be plotted in the x axis. Plot by default plots a frequency Polygon
title(main = "Sales by month")

lines(sales2, type = "o", pch = 22, lty = 2, col = "red")
grid(nx=NA, ny=NULL)
legend("topright", inset = 0.05, c("Sales1", "Sales2"), fill = c("blue", "red"), horiz = TRUE)
```

## Problem 2 

```r
sales <- read.delim('salesdata.txt', header = T)
#sales #to verify that the data has been read
barplot(as.matrix(sales), main="Sales Data", ylab= "Total",beside = T, col=rainbow(5))
```

## Problem 3

```r
fn <- boxplot(sales, col=c("orange","green")) $stats
text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0, cex=.7)
text(0.45, fn[3,1],paste("Median =", fn[3,1]), adj=0, cex=.7)
grid(nx=NA, ny=NULL)
```

## Problem 4

```r
fb1 <- read.csv('FB.csv')
aapl1 <- read.csv('AAPL.csv') 
par(bg = "cornsilk")
plot(aapl1$Adj.Close, col="blue", type="o", ylim=c(160,200), xlab="Days", ylab="Price")
lines(fb1$Adj.Close, type="o", pch=22, lty=2, col="red")
legend("topright", inset=.05, c("Apple","Facebook"), fill=c("blue","red"), horiz=TRUE)
```

## Plotting Histogram

```r
par(bg = "cornsilk")
hist(aapl1$Adj.Close, col=rainbow(8), xlab = "Stock Price of Apple")
```
