## Problem 1

```r
sales1 <- c(12,14,16,29,30,45,19,20,16,19,34,20)
sales2 <- rpois(12,34) #Will return a 12 tupled vector with mean at 34 in Poisson Distribution
par(bg = "cornsilk")

plot(sales1, col = "blue", type = "o", ylim = c(0,100), xlab = "Month", ylab = "Sales") #By default, the count of 
#sales will be plotted in the x axis. Plot by default plots a frequency Polygon
title(main = "Sales by month")

lines(sales2, type = "o", pch = 22, lty = 2, col = "red")
grid(nx=NA, ny=NULL)
legend("topright", inset = 0.05, c("Sales1", "Sales2"), fill = c("blue", "red"), horiz = TRUE)
```
